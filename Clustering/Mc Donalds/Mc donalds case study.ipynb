```python
#Label encoding for categorical


def labelling(x):
    mc[x] = LabelEncoder().fit_transform(mc[x])
    return mc

cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    labelling(i)
mc
```

```python
#converting category variables into array
x = mc.loc[:,cat].values
x
```

```python
#Principal component analysis
pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
#Proportion of Variance 
pca.explained_variance_ratio_
```

```python
#cumulative sum
np.cumsum(pca.explained_variance_ratio_)
```

```python
#Getting the eleven attributes
cat_eleven = mc.loc[:,cat]

cat_eleven
```

```python
# correlation coefficient between original variables and the component

loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = cat_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#Correlation matrix plot for loadings 
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
#seeing weightage of components
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
# get PC scores
pca_scores = PCA().fit_transform(x)
pca_scores
```

```python
#  2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=mc.columns.values, var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(10,5))
```

```python
#Using k-means clustering analysis
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(cat_eleven)
visualizer.show()
```

```python
#K-means clustering 

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(cat_eleven)
mc['cluster_num'] = kmeans.labels_ 
```

```python
#making the clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(cat_eleven)
kmeans
```

```python
#adding to the dataframe
mc['cluster_num'] = kmeans.labels_
```

```python
#getting the countc
Counter(kmeans.labels_)
```

```python
#Visualizing  the clusters
sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```

```python
#presenting in tabular format
crosstab = pd.crosstab(mc['cluster_num'], mc['Like'])


crosstab
```

```python
#Mosaic plot gender vs segment
crosstab_gender =pd.crosstab(mc['cluster_num'],mc['Gender'])
crosstab_gender
```
