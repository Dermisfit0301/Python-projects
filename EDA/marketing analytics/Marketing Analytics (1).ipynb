```python
# Customer retention function
def calculate_customer_retention(data, month_column='order_purchase_timestamp'):
 
    data['month'] = pd.to_datetime(data[month_column]).dt.to_period("M")

    n_customers_per_month = data.groupby('month')['customer_unique_id'].nunique()


    customer_retention = n_customers_per_month.pct_change() * 100


    retention_data = pd.DataFrame({
        'Month': n_customers_per_month.index,
        'Unique Customers': n_customers_per_month.values,
        'Customer Retention (%)': customer_retention.values
    })

    return retention_data
```

```python
#revenue by customer type
def calculate_revenues_by_customer_type(data):
   
    data['customer_type'] = data.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min')
    data['customer_type'] = data['order_purchase_timestamp'] == data['customer_type']
    data['customer_type'] = data['customer_type'].replace({True: 'New', False: 'Existing'})

    # Calculate revenues from 'New' and 'Existing' customers each month
    revenues_by_customer_type = data.groupby(['order_purchase_timestamp', 'customer_type'])['price'].sum().unstack().reset_index()
    revenues_by_customer_type = revenues_by_customer_type.rename(columns={'New': 'Revenue from New Customers', 'Existing': 'Revenue from Existing Customers'})

    return revenues_by_customer_type
```

```python
#Customer Retention on a Month-to-Month Basis
customer_retention = calculate_customer_retention(final)
print('Customer Retention on a Month-to-Month Basis:')
print(customer_retention)
```

```python
#Revenues from Existing/New Customers on a Month-to-Month Basis
revenues_by_customer_type = calculate_revenues_by_customer_type(final)
print('Revenues from Existing/New Customers on a Month-to-Month Basis:')
print(revenues_by_customer_type)
```

```python
# Calculate money spent by customers
customer_money_spent = final.groupby('customer_id').apply(lambda group: (group['order_item_id'] * group['price']).sum())

```

```python
# Classifying customers into spending categories

data = pd.DataFrame({
  'customer_id': customer_money_spent.index,
  'money_spent': customer_money_spent.values
})

quantiles = [0, 0.2, 0.4, 0.5, 0.6, 0.8, 1]


customer_segments = ['Lowest Spending', 'Low Spending', 'Moderate Spending', 'Mid-High Spending', 'High Spending', 'Highest Spending']

data['customer_segment'] = pd.qcut(data['money_spent'], q=quantiles, labels=customer_segments)

print(data[['customer_id', 'customer_segment']])
```

```python
# Calculate money earnt by sellers
seller_money_earnt = final.groupby('seller_id').apply(lambda group: (group['order_item_id'] * group['price']).sum())
```

```python
# Classifying sellers into revenue categories

data = pd.DataFrame({
  'seller_id': seller_money_earnt.index,
  'money_earned': seller_money_earnt.values
})

quantiles = [0, 0.2, 0.4, 0.5, 0.6, 0.8, 1]


customer_segments = ['Lowest earning', 'Low earning', 'Moderate earning', 'Mid-High earning', 'High earning', 'Highest earning']

data['seller_segment'] = pd.qcut(data['money_earned'], q=quantiles, labels=customer_segments)

print(data[['seller_id', 'seller_segment']])
```
